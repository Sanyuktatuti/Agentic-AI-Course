### Sentence Transformers on Hugging Face

Hugging Face sentence-transformers is a Python framework for state-of-the-art sentence, text and image embeddings. One of the embedding models is used in the HuggingFaceEmbeddings class. We have alo added an alias for SentenceTransformerEmbeddings for users who are more familiar with directly using that package.

```python
from langchain_huggingface import HuggingFaceEmbeddings 
embeddings = HuggingFaceEmbeddings(model_name = "all-MiniLM-L6-v2")
```

```python
embeddings
```

```python
text = "This is a text document"
query_result = embeddings.embed_query(text)
query_result
```

```python
len(query_result)

# Much lesser (384) Feature representation that the OpenAI embeddings (3072)
```
