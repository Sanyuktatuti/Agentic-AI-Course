````python
def multiply(a:int, b:int) -> int:
    return a*b
````

````python
multiply(80, 9)
````

If we want to convert the above ```"multiply"``` function into an AI tool we need to import ```"tool"``` function  from ```"langchain.agents"```. 

Via ```"@tool"``` decorator before the functions. \
Note: Function must have a ```docstring``` enclosed between the ```''' '''``` if description not provided. \

Call the tool using the ```invoke``` function and pass the arguments in the dictionary format.


````python
from langchain_core.tools import tool
````

````python
@tool
def multiply(a:int, b:int) -> int:
    '''This tool is for the multiplication of two numbers'''
    return a*b
````

````python
multiply.invoke({"a": 10, "b": 20})
````

````python
multiply.name
````

````python
multiply.description
````

````python
multiply.args
````
